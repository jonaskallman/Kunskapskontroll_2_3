##### Remark on Broadcasting when doing Linear Algebra calculations in Python. 
From the mathematical rules of matrix addition, the operation below (m1 + m2) does not make sense. The reason is that matrix addition requires two matrices of the same size. In Python however, it works due to broadcasting rules in NumPy. So you must be careful when doing Linear Algebra calculations in Python since they do not follow the "mathematical rules". This can however easily be handled by doing some simple programming, for example validating that two matrices have the same shape is easy if you for instance want to add two matrices.

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([1, 1])


# m1 har shape (2, 2)
# m2 har shape (2,)

# Numpy upprepar raden i m2 
#[ [1, 2],     [ [1, 1],     [ [2, 3],
#  [3, 4] ]  +   [1, 1] ]  =   [4, 5] ]

print(m1)
print()
print(m2)
print()
print(m1 + m2)
```

The example below would also not be allowed if following the "mathematical rules" in Linear Algebra. But it works due to broadcasting in NumPy. 

```python
v1 = np.array([1, 2, 3])

# v1 har shape (3,)
# 1  är alltså  [1, 1, 1]

print(v1)
print(v1 + 1)

```

### Vector and matrix algebra

Now you are going to create a function that can be reused every time you add or multiply matrices. The function is created so that we do the addition and multiplication according to the rules of vector- and matrix algebra.

Create a function "add_mult_matrices" that takes two matrices as input arguments (validate that the input are of the type numpy.ndarray by using the isinstance function), a third argument that is either 'add' or 'multiply' that specifies if you want to add or multiply the matrices (validate that the third argument is either 'add' or 'multiply'). When doing matrix addition, validate that the matrices have the same size. When doing matrix multiplication, validate that the sizes conform (i.e. number of columns in the first matrix is equal to the number of rows in the second matrix).

```python
 # Min funktion för att räkna matriser samt validera att de har samma shape etc

import numpy as np

def add_matris(A, B, operation):
    if not isinstance(A, np.ndarray) or not isinstance(B, np.ndarray):
    #if not isinstance(A, np.ndarray) or not isinstance(B, np.ndarray):    
        raise TypeError("Fel! Du måste använda array av typen np.ndarray")

    if operation not in ("add", "multiply"):
        raise ValueError("Måste vara add och multiply")


    if operation == "add":
        if A.shape != B.shape:
            raise ValueError("matriserna måste vara samma shape")
        return A + B

    if operation == "multiply":
        if A.shape[1] != B.shape[0]:
            raise ValueError("matriserna måste vara samma antal kolumner och rader ")
        return A @ B


A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(add_matris(A, B, "add"))
print(add_matris(A, B, "multiply"))



```

```python
#10.2.1
# a till l
import numpy as np

# Definierar matriserna enligt uppgiften
A = np.array([[2, 1, -1],
              [1, -1, 1]])  # A är 2x3

B = np.array([[4, -2, 1],
              [2, -4, -2]])  # B är 2x3

C = np.array([[1, 2],
              [2, 1]])  # C är 2x2

D = np.array([[3, 4],
              [4, 3]])  # D är 2x2

E = np.array([[1],
              [2]])  # E är 2x1

I = np.array([[1, 0],
              [0, 1]])  # I är 2x2 (identitetsmatris)

#a 2A
a_result = 2 * A  # Multiplicerar varje element i A med 2
print("a) 2A =\n", a_result, "\n")  # Skriver ut resultatet

#b B - 2A
b_result = B - 2 * A  # Beräknar 2A och subtraherar elementvis från B
print("b) B - 2A =\n", b_result, "\n")  # Skriver ut resultatet

#c 3C - 2E  (kontrollerar om det är definierat)
if C.shape == E.shape:  # Kollar om C och E har samma dimension (krav för addition/subtraktion)
    c_result = 3 * C - 2 * E  # Skulle räkna om det vore tillåtet
    print("c) 3C - 2E =\n", c_result, "\n")  # Skriver ut resultatet
else:
    print("c) 3C - 2E är inte definierat (olika dimensioner:", C.shape, "och", E.shape, ")\n")  # Förklarar varför

#d 2D - 3C  (kontrollerar om det är definierat)
if D.shape == C.shape:  # Kollar om D och C har samma dimension (krav för subtraktion)
    d_result = 2 * D - 3 * C  # Utför subtraktionen elementvis
    print("d) 2D - 3C =\n", d_result, "\n")  # Skriver ut resultatet
else:
    print("d) 2D - 3C är inte definierat (olika dimensioner)\n")  # Skriver ut felorsak

#e D^T + 2D
e_result = D.T + 2 * D  # Transponerar D och adderar med 2D (elementvis)
print("e) D^T + 2D =\n", e_result, "\n")  # Skriver ut resultatet

#f 2C^T - 2D^T  (kontrollerar om det är definierat)
if C.T.shape == D.T.shape:  # Kollar om transponaten har samma dimension
    f_result = 2 * C.T - 2 * D.T  # Utför subtraktionen elementvis
    print("f) 2C^T - 2D^T =\n", f_result, "\n")  # Skriver ut resultatet
else:
    print("f) 2C^T - 2D^T är inte definierat (olika dimensioner)\n")  # Skriver ut felorsak

#g A^T - B  (kontrollerar om det är definierat)
if A.T.shape == B.shape:  # Kollar om A^T och B har samma dimension
    g_result = A.T - B  # Skulle räkna om det vore tillåtet
    print("g) A^T - B =\n", g_result, "\n")  # Skriver ut resultatet
else:
    print("g) A^T - B är inte definierat (dimensioner:", A.T.shape, "och", B.shape, ")\n")  # Förklarar varför

#h AC  (matrismultiplikation: kolumner i A måste = rader i C)
if A.shape[1] == C.shape[0]:  # Kollar multiplikationsvillkoret
    h_result = A @ C  # Beräknar matrisprodukten
    print("h) AC =\n", h_result, "\n")  # Skriver ut resultatet
else:
    print("h) AC är inte definierat (A:", A.shape, ", C:", C.shape, ")\n")  # Förklarar varför

#i CD  (matrismultiplikation: kolumner i C måste = rader i D)
if C.shape[1] == D.shape[0]:  # Kollar multiplikationsvillkoret
    i_result = C @ D  # Beräknar matrisprodukten
    print("i) CD =\n", i_result, "\n")  # Skriver ut resultatet
else:
    print("i) CD är inte definierat (C:", C.shape, ", D:", D.shape, ")\n")  # Förklarar varför 

# (j) CB  (matrismultiplikation: kolumner i C måste = rader i B)
if C.shape[1] == B.shape[0]:  # Kollar multiplikationsvillkoret (2 == 2)
    j_result = C @ B          # Beräknar matrisprodukten CB
    print("j) CB =\n", j_result, "\n")  # Skriver ut resultatet
else:
    print("j) CB är inte definierat (C:", C.shape, ", B:", B.shape, ")\n")  # Förklarar varför

# (k) CI  (matrismultiplikation: kolumner i C måste = rader i I)
if C.shape[1] == I.shape[0]:  # Kollar multiplikationsvillkoret (2 == 2)
    k_result = C @ I          # Beräknar matrisprodukten CI (ska bli C)
    print("k) CI =\n", k_result, "\n")  # Skriver ut resultatet
else:
    print("k) CI är inte definierat (C:", C.shape, ", I:", I.shape, ")\n")  # Förklarar varför

# (l) AB^T  (matrismultiplikation: kolumner i A måste = rader i B^T)
if A.shape[1] == B.T.shape[0]:  # Kollar multiplikationsvillkoret (3 == 3)
    l_result = A @ B.T          # Beräknar matrisprodukten AB^T
    print("l) AB^T =\n", l_result, "\n")  # Skriver ut resultatet
else:
    print("l) AB^T är inte definierat (A:", A.shape, ", B^T:", B.T.shape, ")\n")  # Förklarar varför



```

Basic indexing creates a view. How can you check if v1 and v2 is a view or copy? If you change the last element in v2 to 123, will the last element in v1 be changed? Why?

```python
v1 = np.arange(4)
v2 = v1[-2:]
print(v1)
print(v2)
```

```python
import numpy as np

# Skapar en array med 4 heltal
v1 = np.arange(4)

# Tar en slice av de två sista elementen 
v2 = v1[-2:]

print("v1:", v1)  # Skriver ut originalarrayen
print("v2:", v2)  # Skriver ut vyn (slicen)
print()

# Kollar om v2 är en view av v1:
# Om v2.base pekar på v1 (eller en array som i sin tur pekar på v1) så är det en view
print("v2.base:", v2.base)  # Om detta inte är None så är v2 en view (delar data med någon annan array)
print("v2 är view av v1?:", v2.base is v1)  # True betyder att v2 direkt är en view av v1
print()

# Ändrar sista elementet i v2
v2[-1] = 123

print("Efter ändring v2[-1] = 123")
print("v1:", v1)  # Om v2 är en view, syns ändringen även här
print("v2:", v2)
```
