### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Pyplot
Man använder funktioner som tex plt.plot(...). Matplotlib håller koll på “nuvarande” figur/axel

Objekt-orienterat
Man skapar objekt med t.ex. fig, ax = plt.subplots() och ritar via ax.plot(...), ax.set_title(...)

Objekt-orienterat (OO) är det rekommenderade eftersom det är tydligare, mer robust och lättare när du har flera figurer.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure
Hela ritytan. Kan innehålla en eller flera grafer


Axes
Själva diagrammet där data ritas. Innehåller x- och y-axel, titel, kurvor m.m.

Axis
En enskild axel (x-axel eller y-axel)

Artist
Allt som faktiskt ritas upp. Tex linjer, punkter, staplar, text, legend
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(4, -4, 50) # skapar en talserie/linje 

y = x ** 2 #gör beräkning


```

```python
#Pylot

plt.plot(x, y) #axlar
plt.title ("y= x2") #Titel 
plt.xlabel("x")
plt.ylabel("y")

plt.show # visar diagram
```

```python
#OO

fig, ax = plt.subplots() # skapar figur och axel
ax.plot(x, y)
ax.set_title("y = x2")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show() #visar diagram
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data för scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data för bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5)) # Skapar en figur

#Scatter plot
axes[0].scatter(random_data_x, random_data_y) # Ritar punktdiagram
axes[0].set_title("Scatter plot") # Titel
axes[0].set_xlabel("X-värden") # X-axelns etikett
axes[0].set_ylabel("Y-värden") # Y-axelns etikett

#Bar plot
axes[1].bar(names, values) # Ritar stapeldiagram
axes[1].set_title("Bar plot") # Titel
axes[1].set_xlabel("Frukt") # X-axelns etikett
axes[1].set_ylabel("Antal") # Y-axelns etikett

# Justerar layout
plt.tight_layout()

# Visar figuren
plt.show()
```
